```python
import matplotlib.pyplot as plt

# Datos para el gráfico de barras
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [10, 24, 36, 48, 50]

# Crear gráfico de barras
plt.figure(figsize=(8, 6))  # Ajustar el tamaño del gráfico
plt.bar(categorias, valores, color='blue')

# Añadir etiquetas
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.title('Gráfico de Barras')

# Mostrar gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

import matplotlib.pyplot as plt

# Datos para el gráfico de torta
etiquetas = ['A', 'B', 'C', 'D', 'E']
valores = [10, 24, 36, 48, 50]
colores = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen']
explode = (0, 0, 0.1, 0, 0)  # Resalta la porción 'C'

# Crear gráfico de torta
plt.figure(figsize=(8, 6))  # Ajustar el tamaño del gráfico
plt.pie(valores, explode=explode, labels=etiquetas, colors=colores,
        autopct='%1.1f%%', shadow=True, startangle=140)

# Asegurar que el gráfico sea circular
plt.axis('equal')

# Añadir título
plt.title('Gráfico de Torta')

# Mostrar gráfico
plt.show()

```

```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt

# Paso 1: Conectar a la base de datos MySQL
conexion = mysql.connector.connect(
    host="192.168.24.229",
    user="monitoreo",
    password="MEC9621nst1tuc10n3s.",
    database="MEC_UEPP"
)

# Paso 2: Leer los datos de la tabla (ajusta la consulta según tus campos)
query = "SELECT departamento, cantidad_servidor FROM MEC_UEPP.bd_tic_infrainterna_v2"
df = pd.read_sql(query, conexion)

# Cerrar la conexión
conexion.close()

# Paso 3: Contar valores agrupados por Campo1
resultado = df.groupby('departamento')['cantidad_servidor'].count().reset_index()

# Paso 4: Graficar los resultados
plt.figure(figsize=(10, 6))
plt.bar(resultado['departamento'], resultado['cantidad_servidor'], color='lightskyblue')
plt.xlabel('Departamento')
plt.ylabel('Conteo de institucion')
plt.title('Conteo de institucion agrupado por departamento')
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para mejor visibilidad
plt.tight_layout()  # Ajustar el layout
plt.show()
```
