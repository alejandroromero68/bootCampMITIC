```python
import pandas as pd

# Cargar datos desde un archivo CSV
df = pd.read_csv(r'C:\varios\datosBootcamp\glosario_informe_15_07_2024_v2.csv', sep=';', encoding='latin1')

# Ver las primeras filas del DataFrame
print(df.head())

# Ver las primeras y últimas filas del DataFrame
print("Primeras filas del DataFrame:")
print(df.head())

print("nÚltimas filas del DataFrame:")
print(df.tail())

# Obtener información general sobre el DataFrame
print("nInformación del DataFrame:")
print(df.info())

# Generar estadísticas descriptivas
print("nEstadísticas descriptivas del DataFrame:")
print(df.describe())

# Inspeccionar los tipos de datos de las columnas
print("nTipos de datos de las columnas:")
print(df.dtypes)

# Contar valores únicos en una columna específica
print("nConteo de valores únicos en la columna 'codigo_categoria':")
print(df['codigo_categoria'].value_counts())

# Obtener valores únicos en una columna específica
print("nValores únicos en la columna 'codigo_descripcion':")
print(df['codigo_descripcion'].unique())

# Obtener el número de valores únicos en una columna específica
print("nNúmero de valores únicos en la columna 'descripcion':")
print(df['descripcion'].nunique())

# Renombrar una columna
df.rename(columns={'categoria': 'categoria_descripcion'}, inplace=True)
print("nDataFrame con columna renombrada:")
print(df.head())

# Cambiar el índice del DataFrame
df.set_index('codigo_glosario', inplace=True)
print("nDataFrame con nuevo índice:")
print(df.head())

# Ordenar el DataFrame por una columna
df_ordenado = df.sort_values(by='descripcion')
print("nDataFrame ordenado por 'descripcion':")
print(df_ordenado.head())

# Ordenar el DataFrame por múltiples columnas
df_ordenado_multi = df.sort_values(by=['descripcion', 'codigo_categoria'], ascending=[True, False])
print("nDataFrame ordenado por 'descripcion' y 'codigo_categoria':")
print(df_ordenado_multi.head())
```
